Graphics to visualize dataset

```python
import matplotlib.pyplot as plt
import pandas as pd
import ast
%matplotlib inline
```

```python
#stock price
df = pd.read_csv("results.csv")
df['Date'] = pd.to_datetime(df['Date'])

plt.title('Tesla Price vs Time')
plt.plot(df["Date"], df["Close"], color='black', label=f"Close Price")
plt.xlabel("Year")
plt.ylabel("Price")
plt.legend()
```

```python
def countList(list):
    return len(ast.literal_eval(list))

# tweets per week
df = pd.read_csv("results.csv")
df['Date'] = pd.to_datetime(df['Date'])
df['TweetCount'] = df['Tweets'].apply(countList)
df['Date'] = pd.to_datetime(df['Date']) - pd.to_timedelta(7, unit='d')

df = df.groupby([pd.Grouper(key='Date', freq='W')])['TweetCount'].sum()
df = df.reset_index()
df = df[(df.TweetCount != 0)]

plt.title('# of Tweets per week')
plt.scatter(df["Date"], df["TweetCount"], color='black', label=f"Tweet count")
plt.xlabel("Date")
plt.ylabel("Tweets per week")
plt.legend()
```

```python
def sumList(list):
    return sum(ast.literal_eval(list))

#Likes per week
df = pd.read_csv("results.csv")
df['Date'] = pd.to_datetime(df['Date'])
df['LikesCount'] = df['Likes'].apply(sumList)
df['Date'] = pd.to_datetime(df['Date']) - pd.to_timedelta(7, unit='d')

df = df.groupby([pd.Grouper(key='Date', freq='W')])['LikesCount'].sum()
df = df.reset_index()
df = df[(df.LikesCount != 0)]

plt.title('# of Likes per week')
plt.plot(df["Date"], df["LikesCount"], color='green', label=f"Likes")
plt.xlabel("Date")
plt.ylabel("Likes per week")
plt.legend()
```

```python
#Retweets per week
df = pd.read_csv("results.csv")
df['Date'] = pd.to_datetime(df['Date'])
df['RetweetsCount'] = df['Retweets'].apply(sumList)
df['Date'] = pd.to_datetime(df['Date']) - pd.to_timedelta(7, unit='d')

df = df.groupby([pd.Grouper(key='Date', freq='W')])['RetweetsCount'].sum()
df = df.reset_index()
df = df[(df.RetweetsCount != 0)]

plt.title('# of Retweets per week')
plt.plot(df["Date"], df["RetweetsCount"], color='blue', label=f"Retweets")
plt.xlabel("Date")
plt.ylabel("Retweets per week")
plt.legend()
```

```python
#Differential per week
df = pd.read_csv("results.csv")
df['Date'] = pd.to_datetime(df['Date'])

plt.title('Differential per Week')
plt.plot(df["Date"], df["Dif"], color='red', label=f"Dif")
plt.xlabel("Date")
plt.ylabel("Differential")
plt.legend()
```

```python
def countList(list):
    return len(ast.literal_eval(list))

#Tweets per month and sotck price
df = pd.read_csv("results.csv")
df['Date'] = pd.to_datetime(df['Date'])
df['TweetCount'] = df['Tweets'].apply(countList)
df['Date'] = pd.to_datetime(df['Date']) - pd.to_timedelta(1, unit='m')

lf = df.groupby([pd.Grouper(key='Date', freq='M')])['TweetCount'].sum()
lf = lf.reset_index()
lf = lf[(lf.TweetCount != 0)]

plt.title('Tweets per month and stock price')
plt.scatter(lf["Date"], lf["TweetCount"], color='black', label=f"Tweets")
plt.plot(df["Date"], df["Close"], color='Green', label=f"Stock Price")
plt.xlabel("Date")
plt.ylabel("Tweets per week & price of stock")
plt.legend()
```
